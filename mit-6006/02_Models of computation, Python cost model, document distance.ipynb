## 1. So, What is an Algorithm?
Algorithm is **Mathematical abstraction** of computer program, or **Computational procedure** to solve a problem.

So, program takes an input which calculates through the computational procedure (or algorithms) to print the output.



### Model of Computation specifies ...
1. what operations an algorithm is allowed
2. cost of each operation
> cost of algorithm = sum of operation costs

## 2. Random Access Machine (RAM)


**RAM** is also referred to as a *Random Access Memory*, and it is modeled by a big array. 

- RAM is usually organized by machine words with its index numbers. 
- in constant time, RAM can
    - load constant words
    - do constant computations
    - store constant words
  
- So, What is the *word* ?
    - assume that word is *w* bits that fits within one memory location
    - its all the data types that fits in there

## 3. Pointer Machine
- It dynamically allocates objects (such as named tuples or objects)
- Object has constant number of fields
    - field is either a word(e.g., *int*) or a pointer to object or null


## 4. Python Model
Python offers both mode of thinking because *list* is basically an array (RAM), and object with constant attributes is basically a pointer machine



### 4-1. list
It's an array in all the other programing languages. 

#### a. `L.append(x)` -> constant time

in ordinary data structures, in order to append an item to a list would require allocation of new memories to make duplicate of the original list plus one for the newly appending item. However, Python uses `table double` to make it done in constant time!

#### b. `L = L1 + L2` concatnation 

List concatnation in python takes following order
```python
L = []
for x in L1:
    L.append(x) # constant time
for x in L2:
    L.append(x) # constant time
```
so the whole operation takes *O(n)*

#### c. `L1.extends(L2)` -> linear time

*L1 += L2* === *L1.append(x) 








## 5. Document Distance Problem  - compute *d(D1, D2)*

Document distance problem has applications in finding similar documents, detecting duplicates such as *Wikepedia mirros* and *Google*, and Plagiarism, and also in web search (D2 = query)



The idea is to define the distance in terms of shared words. 

Think of document *D* as a **vector** where D[w] === number of occerences of word W.
