## Binary Search는?


Binary Search는 어떤 데이터의 컬렉션 안에서 원하는 자료를 찾는 방법중 하나입니다.

### 특징

Binary search는 sorted가 된 array 안에서만 활용이 가능합니다.




만약 숫자 13을 찾고싶다면, 일단 array의 가장 앞의 시작점부터 찾아야겠다고 생각을 할 수 있는데요,
위의 array가 sorted 되어있다는 장점을 이용해서 조금 더 효율성 있게 검색이 가능합니다.

만약 현재 선택하고 있는 숫자가 12라고 하면, sorted 된 상태에서는 13은 12보다 앞에 올 수 없겠죠. 

따라서 12 앞에 있는 숫자들만 검색을 함으로써, 앞에 있는 숫자들 10과 11을 굳이 검색해보지 않아도 됩니다.

반대로, 만약 현재 14에 위치하고 있다면, 13은 14보다 뒤에 올 수 없습니다. 따라서 14 뒤에 있는 숫자들을 제외하고 검색함으로 시간을 아낄 수 있습니다.

### 적용법

**1. Array의 끝에 2개의 포인터를 지정해줍니다**


`array[0]` 의 위치와 `array[len(array)-1]` 의 위치에 포인터를 지정해줍니다.

**2. 두 포인터의 중앙에 오는 값을 찾습니다**

**3. 중앙의 값이 우리가 찾고싶은 값보다 크다면 뒤의 반을 날려버리고, 마지막 포인터를 중앙으로 위치시킵니다**

**4. 중앙의 값이 우리가 찾고싶은 값보다 작다면, 앞의 반을 날리고, 첫번쨰 포인터를 중앙의 인덱스 + 1 에 위치시킵니다**

### 예시
**1. 첫번째 작업**


```
Array 의 끝에 2개의 포인터를 지정해줍니다.

찾고싶은 값 = 15

포인터의 위치:
start = 0
end = 5
```

**2. 중앙의 값 찾기**


```
Array 길이의 중간값을 선택해줍니다
찾고싶은 값 = 15

포인터의 위치:
start = 0
end = 5

선택된 값 = array[(start + end)/2 + 1] 
         == array[(0 + 5)/2 + 1]
         == array[2+1] 
         == array[3]
         
```
