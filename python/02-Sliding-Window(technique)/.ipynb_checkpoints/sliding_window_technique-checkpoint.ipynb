## Sliding Window

Sliding Window는 문제를 optimize 시키는 테크닉중 하나로써, 

integer로 이루어진 N 사이즈의 array에서 연속된 K개의 수들의 합중 최대값을 찾는 방법입니다


### 문제 예시

```
N = 4, K = 2
Arr =[80, -50, 90, 100]

index 위치 0부터 K개의 수의 합을 구하세요
```

### Brute Force Approach

```
1. 일단 index = 0부터 K까지의 수를 current_sum에 더합니다
2. 더해진 current_sum과 현재 저장되어있는 max_sum을 비교하고
3. 둘중 더 큰 값을 max_sum에 대입합니다

function maxSum(arr, k):
    max_sum = NEGATIVE_INFINITY
    n = arr.length
    Loop as i in range 0 < n - k + 1:
        current_sum =0
        Loop as j in range 0 < k:
            current_sum += arr[ i + j ]
        max_sum = max( current_sum, max_sum )
    return max_sum
```

### Sliding Window Approach

```
1. 먼저 작업할 창을 생성
2. max_sum에 창의 값을 입력
3. 루프를 활용해 기존 작업하는 창의
- 가장 앞에 있는 수를 빼고
- 창의 다음에 오는 수를 더합니다
4. 그 다음 기존 max_sum과 새로 생성된 window의 값중 더 큰걸 비교합니다
```

**3. 둘중 더 큰 값 비교**
