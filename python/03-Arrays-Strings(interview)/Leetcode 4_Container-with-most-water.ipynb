# Container with the most water 

### 
Given n non-negative integers 
```
a1, a2, ..., an
``` 
where each represents a point at coordinate (i, ai),

n vertical lines are drawn such that the two endpoints of line i is at `(i, ai)` and `(i,0)`

Find two lines, which together with x-axis forms a container, such that the container contains the most water

### Space Complexity
O(1)

 , variable    

````python
class Solution:
    def maxDifference(self, pillarArray):
        left = 0
        right = len(pillarArray) - 1
        max_area = 0
        
        while (left < right):
            max_area = max(max_area, min(pillarArray[left], pillarArray[right]) * (right - left))
            
            if (pillarArray[left] < pillarArray[right]):
                left += 1
            else:
                right -= 1
        return max_area
    
solution = Solution()
answer = solution.maxDifference([2, 1, 1, 2])
print(answer)
````
