# Longest Substring without Repeating Characters

### 문제

Given a string, find the length of the longsest substring without repeating character

```
Input: abcabcbb
Output: 3
```
`abc`abcbb

처럼 abc까지는 중복되는 낱말이 없지만, 그 이상 가면 결국 반복되기 때문

> Sliding Window Technique을 사용하자!
>1. 시작지점에 포인터 두개를 지정
>```
left = 0
right = 0
>```
>2. 각각의 포인터들이 결국 substring들의 시작점과 마지막을 가리키게 하자
