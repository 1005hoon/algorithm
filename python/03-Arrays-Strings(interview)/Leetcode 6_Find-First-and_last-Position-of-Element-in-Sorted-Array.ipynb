# Find the First and the Last Position of Element in Sorted Array


## 문제
Find the starting and ending position of a given target value

### 주의점
1. Sorted Array => Binary Search!
2. 2 answers to get => first and last position


## 사용 알고리즘
### 1. 첫번째 match 구하기

#### 1-1. 두개의 포인터 생성 (일단 바이너리 서치)


#### 1-2.  left가 right보다 작거나 같을때까지 루프
```

```

#### 1-3. 가운데 값 가져오기 
```

    mid = (left + right) / 2
```    

#### 1-4. mid 값 비교하기
##### 1-4-1. mid 값이 타겟값과 같다면 이게 첫번째 match인지 체크하기
```

while (left <= right):

        # mid의 인덱스가 가장 왼쪽이였거나 || mid보다 앞에 위치한 값이 타겟인지 비교
        if(mid === 0 || nums[mid-1] !== target) :
            return mid
        #만약 같다면 mid-1 을 mid로 지정
        right = mid - 1
```

#### 1-5. 타겟값이 mid 값보다 작은경우 right = mid - 1
```
left = 0
right = len(nums) - 1
while (left <= right):
    mid = (left + right) / 2

        if(mid === 0 || nums[mid-1] !== target) :
```

#### 1-6. 타겟값이 mid 값보다 큰 경우 left = mid + 1
```

        if(mid === 0 || nums[mid-1] !== target) :
```

### 2. 가장 끝의 자리 match 구하기
#### 2-1. 똑같이 2 포인터 세팅

```

right = len(nums) - 1
```

#### 2-2. 똑같이 왼쪽이 오른쪽과 같아질때까지 루프

```
left = 0
right = len(nums) - 1
while (left <= right):
   
```

#### 2-3. match 발견 시 해당 match값이 가장 마지막에 있는 match인지 체크

```
left = 0
right = len(nums) - 1
while (left <= right):
    mid = (left + right) / 2
    if (nums[mid] === target):
        # mid의 인덱스 값이 array의 가장 오른쪽인지 || mid의 오른쪽 값이 타겟인지
        if(mid === len(nums) - 1 || nums[mid + 1] !== target) :
            return mid
        #만약 같다면 mid+1 을 left로 지정
        left = mid - 1
```

#### 2-4. 이후 바이너리 서치 진행
```
left = 0
right = len(nums) - 1
while (left <= right):
    mid = (left + right) / 2
    if (nums[mid] === target):
        # mid의 인덱스 값이 array의 가장 오른쪽인지 || mid의 오른쪽 값이 타겟인지
        if(mid === len(nums) - 1 || nums[mid + 1] !== target) :
            return mid
        #만약 같다면 mid+1 을 left로 지정
        left = mid - 1
    else if(nums[mid] > target):
        right = mid - 1
    else:
        left = mid + 1
```


### Time Complexity: O( Log(N)) 
Binary Search를 이용하기 때문!

### Space Complexity: O(1)


````python
class Solution:
    def getLeft(self, nums, target):
        left = 0
        right = len(nums) - 1
        
        while (left <= right):
            mid = (left + right) // 2
            if (nums[mid] == target):
                if (mid == 0 or mid - 1 >= 0 and nums[mid-1] != target):
                    return mid
                right = mid - 1
            elif (nums[mid] > target):
                right = mid - 1
            else: 
                left = mid + 1
        return -1
    
    def getRight(self, nums, target):
        left = 0
        right = len(nums) - 1
        
        while (left <= right):
            mid = (left + right) // 2
            if (nums[mid] == target):
                if (mid == len(nums) - 1 or mid + 1 < len(nums) and nums[mid+1] != target):
                    return mid
                left = mid + 1
            elif (nums[mid] > target):
                right = mid - 1
            else: 
                left = mid + 1
        return -1
    
    def searchRange(self, nums, target):
        left = self.getLeft(nums, target)
        right = self.getRight(nums, target)
        return [left, right]
````

````python
s = Solution()
answer = s.searchRange([1, 2, 2, 3, 4, 4, 4, 4, 4], 4)
print(answer)
````
